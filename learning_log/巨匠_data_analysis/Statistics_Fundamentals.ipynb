```python
import pandas as pd
import numpy as np
```

## 1. 建立資料集

用一個簡單的例子開始：
記錄10位運動員的年齡和成績

```python
#建立一個簡單的資料集
data = {
    '姓名':['選手A','選手B','選手C','選手D','選手E',
        '選手F','選手G','選手H','選手I','選手J'],
    '年齡':[18,22,25,19,23,21,24,20,26,22],
    '成績':[85,92,78,88,95,71,83,90,76,89]
}

df = pd.DataFrame(data)
print(df)
```

## 2. 基本統計量

描述一筆資料最常用的三個數字：
- 平均數 (mean)：所有數加起來除以個數
- 中位數 (median)：排序後中間那個數
- 標準差 (std)：數值離平均有多遠

```python
# 計算成績的基本統計量
print('平均成績:', df['成績'].mean())
print('中位數:', df['成績'].median())
print('標準差:', df['成績'].std())
print('最高分:', df['成績'].max())
print('最低分:', df['成績'].min())
```

## 3. 一次看所有統計量

```python
# describe() 一次顯示所有統計量
print(df.describe())
```
