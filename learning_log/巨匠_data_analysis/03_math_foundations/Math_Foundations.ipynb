## 1. 向量 (Vector)

向量就是一排數字，代表空間中的一個點或方向。

運動員例子：
選手A的特徵向量 = [訓練時數, 體重, 年齡]
                = [25, 70, 22]

向量的兩個重要概念：
- 大小（magnitude）：向量的長度
- 方向：向量指向哪裡

```python
# 向量基本運算
a = np.array([25, 70, 22])   # 選手A
b = np.array([20, 65, 25])   # 選手B

# 大小（magnitude）
mag_a = np.linalg.norm(a)
print(f'選手A特徵向量大小：{mag_a:.2f}')

# 向量加法
print(f'向量相加：{a + b}')

# 內積（dot product）— 衡量相似度
dot = np.dot(a, b)
print(f'內積：{dot}')

# 餘弦相似度 — 兩位選手有多像
cos_sim = dot / (np.linalg.norm(a) * np.linalg.norm(b))
print(f'餘弦相似度：{cos_sim:.4f}')
print('（越接近1代表兩位選手越相似）')
```

```python
# 視覺化向量（2D）
v1 = np.array([3, 2])
v2 = np.array([1, 4])

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy',
           scale=1, color='steelblue', label='v1=[3,2]')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy',
           scale=1, color='coral', label='v2=[1,4]')
plt.quiver(0, 0, v1[0]+v2[0], v1[1]+v2[1],
           angles='xy', scale_units='xy', scale=1,
           color='green', label=f'v1+v2={v1+v2}')

plt.xlim(-1, 6)
plt.ylim(-1, 7)
plt.grid(alpha=0.3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title('Vector Addition')
plt.show()
```

## 2. 矩陣 (Matrix)

矩陣是二維的數字陣列。

深度學習中，每一層神經網路的計算就是：
輸出 = 矩陣乘法（權重矩陣 × 輸入向量）+ 偏差值

這就是為什麼線性代數是深度學習的基礎！

```python
# 矩陣基本運算
A = np.array([[1, 2, 3],
              [4, 5, 6]])   # 2×3 矩陣

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])    # 3×2 矩陣

# 矩陣乘法
C = A @ B
print('A @ B =')
print(C)
print(f'結果形狀：{C.shape}')

# 轉置
print(f'\nA 的轉置：')
print(A.T)

# 特徵值（深度學習分析用）
M = np.array([[4, 2],
              [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(M)
print(f'\n特徵值：{eigenvalues}')
```

## 3. 線性代數在深度學習的應用

神經網路一層的計算：

輸入：x = [x1, x2, x3]（特徵向量）
權重：W（矩陣）
輸出：y = W × x + b

這就是矩陣乘法！
深度學習 = 很多層矩陣乘法 + 激活函數

```python
# 模擬一層神經網路的計算
np.random.seed(42)

# 輸入：3個特徵
x = np.array([25.0, 70.0, 22.0])   # 訓練時數、體重、年齡

# 權重矩陣（2個神經元，3個輸入）
W = np.random.randn(2, 3)

# 偏差值
b = np.random.randn(2)

# 前向計算
z = W @ x + b

# 激活函數（ReLU：負數變0）
def relu(x):
    return np.maximum(0, x)

output = relu(z)

print(f'輸入：{x}')
print(f'權重矩陣：\n{W.round(3)}')
print(f'線性輸出：{z.round(3)}')
print(f'ReLU後：{output.round(3)}')
```
