```python
import pandas as pd
import numpy as np
```

## 1. 建立資料集

用一個簡單的例子開始：
記錄10位運動員的年齡和成績

```python
#建立一個簡單的資料集
data = {
    '姓名':['選手A','選手B','選手C','選手D','選手E',
        '選手F','選手G','選手H','選手I','選手J'],
    '年齡':[18,22,25,19,23,21,24,20,26,22],
    '成績':[85,92,78,88,95,71,83,90,76,89]
}

df = pd.DataFrame(data)
print(df)
```

## 2. 基本統計量

描述一筆資料最常用的三個數字：
- 平均數 (mean)：所有數加起來除以個數
- 中位數 (median)：排序後中間那個數
- 標準差 (std)：數值離平均有多遠

```python
# 計算成績的基本統計量
print('平均成績:', df['成績'].mean())
print('中位數:', df['成績'].median())
print('標準差:', df['成績'].std())
print('最高分:', df['成績'].max())
print('最低分:', df['成績'].min())
```

## 3. 一次看所有統計量

```python
# describe() 一次顯示所有統計量
print(df.describe())
```

## 5. 資料視覺化

用圖表讓數據更直觀。
- 長條圖：比較各選手成績
- 盒鬚圖：看分布與離群值

```python
import matplotlib.pyplot as plt

# 長條圖：各選手成績
plt.figure(figsize=(10, 4))
plt.bar(df['姓名'], df['成績'], color='steelblue')
plt.title('各選手成績比較')
plt.xlabel('姓名')
plt.ylabel('成績')
plt.axhline(df['成績'].mean(), color='red',
            linestyle='--', label='平均線')
plt.legend()
plt.show()
```

```python
# 盒鬚圖：看成績分布
plt.figure(figsize=(5, 4))
plt.boxplot(df['成績'], labels=['成績'])
plt.title('成績分布盒鬚圖')
plt.ylabel('分數')
plt.show()
```

## 6. 盒鬚圖怎麼看

- 箱子中間的線 = 中位數
- 箱子上下邊 = 第75和第25百分位
- 鬚線末端 = 最大/最小值（排除離群值）
- 超出鬚線的點 = 離群值（異常表現）

運動科學應用：
用盒鬚圖可以快速看出哪位選手的表現異常，
判斷是受傷、狀態特別好、還是測量錯誤。

```python
# 年齡與成績的關係
plt.figure(figsize=(6, 4))
plt.scatter(df['年齡'], df['成績'], color='steelblue', s=80)
plt.title('Age vs Performance')
plt.xlabel('Age')
plt.ylabel('Score')

# 標上每個點的選手名
for i, row in df.iterrows():
    plt.annotate(row['姓名'], (row['年齡'], row['成績']),
                 textcoords='offset points', xytext=(5, 5))
plt.show()

# 計算相關係數
corr = df['年齡'].corr(df['成績'])
print(f'年齡與成績的相關係數：{corr:.3f}')
```
